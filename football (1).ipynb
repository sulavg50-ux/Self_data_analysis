```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa=pd.read_csv('fifa.csv')
fifa
```

```python
fifa.head()
```

```python
fifa.shape
```

```python
fifa.drop_duplicates(inplace=True)
print(fifa.isnull().sum())
```

```python
fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(6,6))
plt.bar(fifa['nationality'].value_counts()[0:5].keys(),
        fifa['nationality'].value_counts()[0:5],
        color=('red','blue','green','grey','yellow'))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(fifa['nationality'].value_counts(),
         fifa['nationality'].value_counts())
plt.title('Line plot for Nationality')
plt.xlabel('Number')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(y="nationality", data=fifa,
              order=fifa['nationality'].value_counts().index[:10])

plt.title("Top 10 Nationalities")
plt.show()
```

```python
#short name with their long name
salary=fifa[['short_name','wage_eur']]
salary
```

```python
salary.head()
```

```python
salary=salary.sort_values(by=['wage_eur'],ascending=False)
salary.head(10)
```

```python
top5=salary.head(10)
plt.bar(top5['short_name'],top5['wage_eur'],color='blue')
plt.title('Top 10 player by wages')
plt.xlabel('Name')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(top5['wage_eur'],
        labels=top5['short_name'],
        autopct='%1.1f%%')
plt.show()
```

```python
por=fifa[fifa['nationality']=="Portugal"]
```

```python
por.head()
```

```python
# shooter in portugal
shoot=por[['short_name','shooting']]
shoot
```

```python
#top 5 best shooter
shoot=shoot.sort_values(by=['shooting'],ascending=False)
shoot.head()
```

```python
#barplot
s1=shoot.head(10)
plt.figure(figsize=(6,6))
plt.bar(s1['short_name'],
        s1["shooting"],
        color='blue')
plt.title('Top 5 shooter')
plt.xticks(rotation=60)
plt.show()
```

```python
#defender
defend=por[['short_name','defending']]
defend.head(10)
    
```

```python
#top 5 defender
defend=defend.sort_values(by=['defending'],ascending=False)
```

```python
defend.head(5)
```

```python
#acc to height
hei=por[['short_name','height_cm']]
hei.head()
```

```python
#sorting
hei=hei.sort_values(by=['height_cm'],ascending=False)
hei.head(10)
```

```python
h1=hei.head(10)
plt.figure(figsize=(6,6))
plt.bar(h1['short_name'],h1['height_cm'])
plt.title('Acc. to Height')
plt.xlabel("Height in cm")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(fifa['age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Counts')
plt.show()
```

```python
#overall vs wages
plt.scatter(fifa['overall'],fifa['wage_eur'])
plt.xlabel("Overall Rating")
plt.ylabel("Wage")
plt.show()
```

```python
fifa[['overall','pace','shooting','passing','dribbling','defending']].corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    fifa[['overall','pace','shooting','passing','dribbling','defending']].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
